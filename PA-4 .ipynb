```python
import pandas as pd
```

Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]

```python
# 1A Initial Data
data = {
    'Name': ['S1', 'S2', 'S3', 'S8', 'S17', 'S30'],
    'GEAS': [75, 65, 80, 64, 79, 57],
    'Electronics': [89, 73, 87, 81, 73, 81]
}

# Create DataFrame
bd = pd.DataFrame(data)

# Applying the filter for specific rows based on the GEAS and Electronics scores
# Here we replicate the rows shown in your image: S1, S8, S30
filtered_df = bd.loc[bd['Name'].isin(['S1', 'S8', 'S30']), ['Name', 'GEAS', 'Electronics']]

# Display the filtered DataFrame
print(filtered_df)
```

Filename: Mindy = [“Name”, “Track”, “Electronics”, “Average >=55”]

```python
# Create a DataFrame 'bd' from the provided data
bd = pd.DataFrame(data)

# Add a new column 'Hometown' to the DataFrame with specified values
bd['Hometown'] = ['Luzon', 'Mindanao', 'Mindanao', 'Mindanao', 'Mindanao', 'Mindanao']

# Add a new column 'Gender' to the DataFrame with specified values
bd['Gender'] = ['Male', 'Female', 'Female', 'Female', 'Female', 'Female']

# Calculate the average of the specified subjects and store it in a new column 'Average'
bd['Average'] = bd[['Math', 'Electronics', 'GEAS']].mean(axis=1)

# Filter the DataFrame for females from Mindanao with an average score of 55 or higher
mindy_df = bd.loc[
    (bd['Hometown'] == 'Mindanao') & (bd['Gender'] == 'Female') & (bd['Average'] >= 55),
    ['Name', 'Track', 'Electronics', 'Average']  # Select specific columns to include in the new DataFrame
]

# Print the filtered DataFrame without the index
print("\nMindy DataFrame:")
print(mindy_df.to_string(index=False))
```

```python
# Load your dataset
df = pd.read_excel('board2.xlsx')
df
```

```python
# Calculate the average grade for each student
df['Average Grade'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

# Plot 1: Average Grade by Track
plt.figure(figsize=(10, 6))
sns.barplot(x='Track', y='Average Grade', data=df)
plt.title('Average Grade by Track')
plt.show()

# Plot 2: Average Grade by Gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Average Grade', data=df)
plt.title('Average Grade by Gender')
plt.show()

# Plot 3: Average Grade by Hometown
plt.figure(figsize=(10, 6))
sns.barplot(x='Hometown', y='Average Grade', data=df)
plt.title('Average Grade by Hometown')
plt.show()

```
